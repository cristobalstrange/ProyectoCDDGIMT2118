```python
import zipfile 
import os 
import fiona 
import pandas as pd 
import geopandas as gpd 
from bs4 import BeautifulSoup 

fiona.drvsupport.supported_drivers['libkml'] = 'rw' 
fiona.drvsupport.supported_drivers['LIBKML'] = 'rw'
```

```python
import shutil 
import os

def save_kmz(kmz_file_path, folder_name):
    if os.path.isdir(folder_name):
        shutil.rmtree(folder_name)
    os.mkdir(folder_name) 
    with zipfile.ZipFile(kmz_file_path, "r") as kmz:
        kmz.extractall(folder_name)

save_kmz("ACS X REGION.kmz", '1')
save_kmz("ACS XI REGION.kmz", '2')
save_kmz("1 BARRIOS PROXMTE EN DESCANSO.kmz", '3')
save_kmz("2 EDUARDO ALMONACID.kmz", '4')
save_kmz("3 INGRID SALINAS.kmz", '5')
save_kmz("4 JAVIER DIAZ.kmz", '6')
```

```python
def read_data(file_path, mode):
    gdf_list = []
    for layer in fiona.listlayers(file_path) :    
        gdf = gpd.read_file(file_path, driver='LIBKML', layer=layer)
        gdf_list.append(gdf)
    if mode == '1':
        return gpd.GeoDataFrame(pd.concat(gdf_list, ignore_index=True))[['Name', 'geometry']].to_crs(4326)
    else:
        return gpd.GeoDataFrame(pd.concat(gdf_list, ignore_index=True))[:24][['Name', 'geometry']].to_crs(4326)
    

gdf_zone = []
gdf_center = []
gdf_all = []

gdf_zone.append(read_data('1/doc.kml', '1'))
gdf_zone.append(read_data('2/doc.kml', '1'))
gdf_descanso = (read_data('3/doc.kml', '2'))
gdf_center.append(read_data('4/doc.kml', '2'))
gdf_center.append(read_data('5/doc.kml', '2'))
gdf_center.append(read_data('6/doc.kml', '2'))

gdf_all.append(read_data('1/doc.kml', '1'))
gdf_all.append(read_data('2/doc.kml', '1'))
# gdf_all.append(read_data('3/doc.kml', '2'))
gdf_all.append(read_data('4/doc.kml', '2'))
gdf_all.append(read_data('5/doc.kml', '2'))
gdf_all.append(read_data('6/doc.kml', '2'))

gdf_zone = pd.concat(gdf_zone)
gdf_center = pd.concat(gdf_center)
gdf_all = pd.concat(gdf_all)


gdf_all.explore()
```
